# hPz for pion and muon : pimu decay - 601MeV.

```python
import ROOT
from IPython.display import Image, display

f = ROOT.TFile("root_files/hepmc_601MeV_pimu.root", "READ")
# g = ROOT.TFile("root_files/hepmc_601MeV_mumu.root", "READ")

folder1 = f.Get("generator")
# folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)

pi = folder1.Get("hPzPi")
mu = folder1.Get("hPzLep")

mu.SetLineColor(ROOT.kRed)
pi.SetLineColor(ROOT.kBlue)

mu.SetLineWidth(2)
pi.SetLineWidth(2)

mu.SetTitle("Muon and Pion z-Momentum Distribution")
pi.SetTitle("Muon and Pion z-Momentum Distribution")

mu.Draw()
pi.Draw("same")
legend = ROOT.TLegend(0.7,0.7,0.9,0.9)
legend.AddEntry(mu, r"$\mu$")
legend.AddEntry(pi, r"$\pi$")
legend.AddEntry(mu, "Mean: %.2f" % mu.GetMean(), "l")
legend.AddEntry(pi, "Mean: %.2f" % pi.GetMean(), "l")

legend.Draw()

canvas.SaveAs("plots/ETot_601_pimu.png")
display(Image(filename="plots/ETot_601_pimu.png"))
```

# Energy for pion and muon : pimu decay - 601MeV.

```python
import ROOT

f = ROOT.TFile("root_files/hepmc_601MeV_pimu.root", "READ")
# g = ROOT.TFile("root_files/hepmc_601MeV_mumu.root", "READ")

folder1 = f.Get("generator")
# folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)


hE1 = folder1.Get("hEnergyLep")
hE1.SetLineColor(ROOT.kBlue)
hE1.SetLineWidth(2)
hE1.Draw()

hE2 = folder1.Get("hEnergyPi")
hE2.SetLineColor(ROOT.kRed)
hE2.SetLineWidth(2)
hE2.Draw("same")



hE1.SetTitle("Muon and Pion Energy Distribution")
hE2.SetTitle("Muon and Pion Energy Distribution")

hE1.SetYTitle("Counts")
hE2.SetYTitle("Counts")

mean1 = hE1.GetMean()
mean2 = hE2.GetMean()

legend = ROOT.TLegend(0.7,0.7,0.97,0.97)
legend.AddEntry(hE1, r"$\mu$")
legend.AddEntry(hE2, r"$\pi$")
legend.AddEntry(hE1, "Mean: %.2f" % mean1, "l")
legend.AddEntry(hE2, "Mean: %.2f" % mean2, "l")

legend.Draw()

canvas.Draw()
```
