```python
# open new TFile to read from
f = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_truthhists_mupi_hdgeom.root", "READ")
#get folder geometry
folder = f.Get("generator")
canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900, 550)

#plot the TObject hPx in the folder

hPx = folder.Get("hPx")
hPx.Draw()
hPx.SetLineColor(ROOT.kRed)
hPx.SetFillColor(ROOT.kRed)
hPx.SetFillStyle(3004)
hPx.SetLineWidth(2)
hPx.SetMarkerStyle(21)
hPx.SetMarkerColor(ROOT.kBlue)
hPx.SetMarkerSize(1.5)
hPx.SetTitle("hPx")
canvas.SetTitle("hPx")
canvas.Draw()
```

```python
# open new TFile to read from
f = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_truthhists_mupi_hdgeom.root", "READ")
folder = f.Get("generator")
canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 800, 800)

#plot the TObject hPx in the folder

hPx = folder.Get("hPx")
hPy = folder.Get("hPy")

# Create a 2D histogram
nbinsx = hPx.GetNbinsX()
nbinsy = hPy.GetNbinsX()
minx = hPx.GetXaxis().GetXmin()
maxx = hPx.GetXaxis().GetXmax()
miny = hPy.GetXaxis().GetXmin()
maxy = hPy.GetXaxis().GetXmax()

h2D = ROOT.TH2F("h2D", "hPx vs hPy; hPx; hPy", nbinsx, minx, maxx, nbinsy, miny, maxy)

# Fill the 2D histogram
for i in range(1, nbinsx+1):
    for j in range(1, nbinsy+1):
        x_value = hPx.GetBinContent(i)
        y_value = hPy.GetBinContent(j)
        h2D.Fill(x_value, y_value)

# Draw the 2D histogram
h2D.Draw("COLZ")


canvas.Draw()
```

```python
# open new TFile to read from
f = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_truthhists_mupi_hdgeom.root", "READ")
folder = f.Get("generator")
canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 800, 800)

#plot the TObject hPx in the folder

hx = folder.Get("hx")
hy = folder.Get("hy")
hz = folder.Get("hz")

# Create a 3D histogram
nbinsx = hx.GetNbinsX()
nbinsy = hy.GetNbinsX()
nbinsz = hz.GetNbinsX()
minx = hx.GetXaxis().GetXmin()
maxx = hx.GetXaxis().GetXmax()
miny = hy.GetXaxis().GetXmin()
maxy = hy.GetXaxis().GetXmax()
minz = hz.GetXaxis().GetXmin()
maxz = hz.GetXaxis().GetXmax()

h3D = ROOT.TH3F("h3D", "hx vs hy vs hz; hx; hy; hz", nbinsx, minx, maxx, nbinsy, miny, maxy, nbinsz, minz, maxz)

# Fill the 3D histogram
for i in range(1, nbinsx+1):
    for j in range(1, nbinsy+1):
        for k in range(1, nbinsz+1):
            x_value = hx.GetBinContent(i)
            y_value = hy.GetBinContent(j)
            z_value = hz.GetBinContent(k)
            h3D.Fill(x_value, y_value, z_value)

# Draw the 3D histogram
h3D.Draw("BOX")

canvas.Draw()
```
