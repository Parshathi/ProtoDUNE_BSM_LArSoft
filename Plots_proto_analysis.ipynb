# Different decays at 601 MeV

## hPz

```python
import ROOT

f = ROOT.TFile("root_files/Protodune_601MeV_pimu.root", "READ")
g = ROOT.TFile("root_files/Protodune_601MeV_mumu.root", "READ")

folder1 = f.Get("generator")
folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)

pimu = folder1.Get("hPz")
mumu = folder2.Get("hPz")

mumu.SetLineColor(ROOT.kRed)
pimu.SetLineColor(ROOT.kBlue)

mumu.SetLineWidth(2)
pimu.SetLineWidth(2)

mumu.SetTitle("Muon-Muon and Pion-Muon z-Momentum Distribution")
pimu.SetTitle("Muon-Muon and Pion-Muon z-Momentum Distribution")

mumu.Draw()
pimu.Draw("same")
legend = ROOT.TLegend(0.7,0.7,0.9,0.9)
legend.AddEntry(mumu, r"$\mu- \mu$")
legend.AddEntry(pimu, r"$\pi - \mu$")
legend.AddEntry(mumu, "Mean: %.2f" % mumu.GetMean(), "l")
legend.AddEntry(pimu, "Mean: %.2f" % pimu.GetMean(), "l")

legend.Draw()

canvas.Draw()
```
