# Plot of Track Length of pion and muon: 601MeV

```python
file = ROOT.TFile("root_files/proto_hist_7July_1518_601MeV_pimu.root", "READ")
dir = file.Get("ana")
tree_mu = dir.Get("tracklengthpimu_muon")
tree_pi = dir.Get("tracklengthpimu_pion")


branch_mu = tree_mu.GetBranch("TLMu")
branch_pi = tree_pi.GetBranch("TLPi")



c=True
c = ROOT.TCanvas("c", "c", 900, 600)
c.cd()


h_mu = ROOT.TH1F("h_mu", "h_mu", 50, 0, 2500)
h_pi = ROOT.TH1F("h_pi", "h_pi", 50, 0, 2500)

tree_mu.Draw("TLMu>>h_mu")
tree_pi.Draw("TLPi>>h_pi")

h_mu.SetLineColor(ROOT.kBlue)
h_pi.SetLineColor(ROOT.kRed)

h_mu.Draw()
h_pi.Draw("same")

h_mu.SetStats(0)
h_pi.SetStats(0)



h_mu.GetXaxis().SetTitle("Track Length (cm)")
h_mu.GetYaxis().SetTitle("Number of Events")


h_mu.SetTitle("Track Length of Muon and Pion:601MeV")
h_pi.SetTitle("Track Length of Muon and Pion:601MeV")

#mean
mean_mu = h_mu.GetMean()
mean_pi = h_pi.GetMean()

#add legend
legend = ROOT.TLegend(0.6, 0.6, 0.9, 0.9)
legend.AddEntry(h_mu, "Muon", "l")
legend.AddEntry(h_pi, "Pion", "l")
legend.AddEntry(h_mu, "Mean of Muon: "+str(mean_mu), "l")
legend.AddEntry(h_pi, "Mean of Pion: "+str(mean_pi), "l")
legend.Draw()



#print
print("Mean of Muon: ", mean_mu)
print("Mean of Pion: ", mean_pi)


c.SaveAs("plots/plot_tl_2.png")


#display the image in the ipynb
from IPython.display import Image
Image("plots/plot_tl_2.png")


```

# Comparison of all the 3 masses: fTrackLength

```python

file1 = ROOT.TFile("root_files/proto_hist_7July_1518_601MeV_pimu.root", "READ")
file2 = ROOT.TFile("root_files/proto_hist_8July_1433_1003MeV_pimu.root", "READ")
file3 = ROOT.TFile("root_files/proto_hist_8July_1405_1405MeV_pimu.root", "READ")

dir1 = file1.Get("ana")
dir2 = file2.Get("ana")
dir3 = file3.Get("ana")

tree_mu_1 = dir1.Get("tracklengthpimu_muon")
tree_pi_1 = dir1.Get("tracklengthpimu_pion")
tree_mu_2 = dir2.Get("tracklengthpimu_muon")
tree_pi_2 = dir2.Get("tracklengthpimu_pion")
tree_mu_3 = dir3.Get("tracklengthpimu_muon")
tree_pi_3 = dir3.Get("tracklengthpimu_pion")


branch_mu_1 = tree_mu_1.GetBranch("TLMu")
branch_pi_1 = tree_pi_1.GetBranch("TLPi")
branch_mu_2 = tree_mu_2.GetBranch("TLMu")
branch_pi_2 = tree_pi_2.GetBranch("TLPi")
branch_mu_3 = tree_mu_3.GetBranch("TLMu")
branch_pi_3 = tree_pi_3.GetBranch("TLPi")


c=True
c = ROOT.TCanvas("c", "c", 900, 600)
c.cd()

h_mu_1 = ROOT.TH1F("h_mu_1", "h_mu_1", 50, 0, 2500)
h_pi_1 = ROOT.TH1F("h_pi_1", "h_pi_1", 50, 0, 2500)
h_mu_2 = ROOT.TH1F("h_mu_2", "h_mu_2", 50, 0, 2500)
h_pi_2 = ROOT.TH1F("h_pi_2", "h_pi_2", 50, 0, 2500)
h_mu_3 = ROOT.TH1F("h_mu_3", "h_mu_3", 50, 0, 2500)
h_pi_3 = ROOT.TH1F("h_pi_3", "h_pi_3", 50, 0, 2500)

tree_mu_1.Draw("TLMu>>h_mu_1")
tree_pi_1.Draw("TLPi>>h_pi_1")
tree_mu_2.Draw("TLMu>>h_mu_2")
tree_pi_2.Draw("TLPi>>h_pi_2")
tree_mu_3.Draw("TLMu>>h_mu_3")
tree_pi_3.Draw("TLPi>>h_pi_3")

h_mu_1.SetLineColor(ROOT.kBlue)
h_pi_1.SetLineColor(ROOT.kRed)
h_mu_2.SetLineColor(ROOT.kGreen)
h_pi_2.SetLineColor(ROOT.kYellow)
h_mu_3.SetLineColor(ROOT.kBlack)
h_pi_3.SetLineColor(ROOT.kOrange)

h_mu_1.Draw()
h_pi_1.Draw("same")
h_mu_2.Draw("same")
h_pi_2.Draw("same")
h_mu_3.Draw("same")
h_pi_3.Draw("same")

h_mu_1.SetStats(0)
h_pi_1.SetStats(0)
h_mu_2.SetStats(0)
h_pi_2.SetStats(0)
h_mu_3.SetStats(0)
h_pi_3.SetStats(0)



h_mu_1.GetXaxis().SetTitle("Track Length (cm)")
h_mu_1.GetYaxis().SetTitle("Number of Events")



h_mu_1.SetTitle("Track Length of Muon and Pion")
h_pi_1.SetTitle("Track Length of Muon and Pion")

#linewidth
h_mu_1.SetLineWidth(2)
h_pi_1.SetLineWidth(2)
h_mu_2.SetLineWidth(2)
h_pi_2.SetLineWidth(2)
h_mu_3.SetLineWidth(2)
h_pi_3.SetLineWidth(2)

#mean
mean_mu_1 = h_mu_1.GetMean()
mean_pi_1 = h_pi_1.GetMean()
mean_mu_2 = h_mu_2.GetMean()
mean_pi_2 = h_pi_2.GetMean()
mean_mu_3 = h_mu_3.GetMean()
mean_pi_3 = h_pi_3.GetMean()

#add legend
legend = ROOT.TLegend(0.5, 0.5, 0.9, 0.9)
# legend.AddEntry(h_mu_1, "Muon", "l")
# legend.AddEntry(h_pi_1, "Pion", "l")
legend.AddEntry(h_mu_1, "Mean of Muon 601MeV: "+str(mean_mu_1), "l")
legend.AddEntry(h_pi_1, "Mean of Pion 601MeV: "+str(mean_pi_1), "l")
legend.AddEntry(h_mu_2, "Mean of Muon 1003MeV: "+str(mean_mu_2), "l")
legend.AddEntry(h_pi_2, "Mean of Pion 1003MeV: "+str(mean_pi_2), "l")
legend.AddEntry(h_mu_3, "Mean of Muon 1405MeV: "+str(mean_mu_3), "l")
legend.AddEntry(h_pi_3, "Mean of Pion 1405MeV: "+str(mean_pi_3), "l")
legend.Draw()




c.SaveAs("plots/plot_tl_masses.png")

#display the image in the ipynb
from IPython.display import Image
Image("plots/plot_tl_masses.png")
```
