```python
import ROOT
```

# Plot of Track Length of pion and muon

```python
file = ROOT.TFile("root_files/proto_hist_7July_1518_601MeV_pimu.root", "READ")
dir = file.Get("ana")
tree_mu = dir.Get("tracklengthpimu_muon")
tree_pi = dir.Get("tracklengthpimu_pion")


branch_mu = tree_mu.GetBranch("TLMu")
branch_pi = tree_pi.GetBranch("TLPi")



c=True
c = ROOT.TCanvas("c", "c", 900, 600)
c.cd()


h_mu = ROOT.TH1F("h_mu", "h_mu", 50, 0, 2500)
h_pi = ROOT.TH1F("h_pi", "h_pi", 50, 0, 2500)

tree_mu.Draw("TLMu>>h_mu")
tree_pi.Draw("TLPi>>h_pi")

h_mu.SetLineColor(ROOT.kBlue)
h_pi.SetLineColor(ROOT.kRed)

h_mu.Draw()
h_pi.Draw("same")

h_mu.SetStats(0)
h_pi.SetStats(0)



h_mu.GetXaxis().SetTitle("Track Length (cm)")
h_mu.GetYaxis().SetTitle("Number of Events")


h_mu.SetTitle("Track Length of Muon and Pion")
h_pi.SetTitle("Track Length of Muon and Pion")



#add legend
legend = ROOT.TLegend(0.8, 0.8, 0.9, 0.9)
legend.AddEntry(h_mu, "Muon", "l")
legend.AddEntry(h_pi, "Pion", "l")
legend.Draw()




c.SaveAs("plots/plot_tl_2.png")


#display the image in the ipynb
from IPython.display import Image
Image("plots/plot_tl_2.png")


```

# Plot Track Length to exit of pion and muon

```python
file = ROOT.TFile("root_files/proto_hist_7July_1518_601MeV_pimu.root", "READ")
dir = file.Get("ana")
tree_mu = dir.Get("tracklengthtoexit_pimu_muon")
tree_pi = dir.Get("tracklengthtoexit_pimu_pion")


branch_mu = tree_mu.GetBranch("TLtoExitMu")
branch_pi = tree_pi.GetBranch("TLtoExitPi")



c=True
c = ROOT.TCanvas("c", "c", 900, 600)
c.cd()

del h_mu
del h_pi
h_mu = ROOT.TH1F("h_mu", "h_mu", 50, 0, 1000)
h_pi = ROOT.TH1F("h_pi", "h_pi", 50, 0, 1000)

tree_mu.Draw("TLtoExitMu>>h_mu")
tree_pi.Draw("TLtoExitPi>>h_pi")

h_mu.SetLineColor(ROOT.kBlue)
h_pi.SetLineColor(ROOT.kRed)

h_mu.Draw()
h_pi.Draw("same")

h_mu.SetStats(0)
h_pi.SetStats(0)



h_mu.GetXaxis().SetTitle("Track Length to Exit (cm)")
h_mu.GetYaxis().SetTitle("Number of Events")


h_mu.SetTitle("Track Length to Exit of Muon and Pion")
h_pi.SetTitle("Track Length to Exit of Muon and Pion")



#add legend
legend = ROOT.TLegend(0.8, 0.8, 0.9, 0.9)
legend.AddEntry(h_mu, "Muon", "l")
legend.AddEntry(h_pi, "Pion", "l")
legend.Draw()




c.SaveAs("plots/plot_tl_2.png")


#display the image in the ipynb
from IPython.display import Image
Image("plots/plot_tl_2.png")


```
