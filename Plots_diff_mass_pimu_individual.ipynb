# hPx for different masses: Pion

```python
import ROOT 
from IPython.display import Image, display
# open new TFile to read from
f = ROOT.TFile("root_files/hepmc_601MeV_pimu.root", "READ")
g = ROOT.TFile("root_files/hepmc_1003MeV_pimu.root", "READ")
h = ROOT.TFile("root_files/hepmc_1405MeV_pimu.root", "READ")
folder1 = f.Get("generator")
folder2 = g.Get("generator")
folder3 = h.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)

#plot the TObject hPx in the folder

hPx1 = folder1.Get("hPxPi")
hPx1.SetLineColor(ROOT.kBlue)
hPx1.SetLineWidth(2)
hPx1.Draw()

hPx2 = folder2.Get("hPxPi")
hPx2.SetLineColor(ROOT.kRed)
hPx2.SetLineWidth(2)
hPx2.Draw("same")

hPx3 = folder3.Get("hPxPi")
hPx3.SetLineColor(ROOT.kGreen)
hPx3.SetLineWidth(2)
hPx3.Draw("same")



ROOT.gStyle.SetOptStat(0)

hPx1.SetTitle("hPx for 3 masses " + r"$\pi$")
hPx2.SetTitle("hPx for 3 masses " + r"$\pi$")
hPx3.SetTitle("hPx for 3 masses " + r"$\pi$")

hPx1.SetYTitle("Counts")
hPx2.SetYTitle("Counts")
hPx3.SetYTitle("Counts")

mean1 = hPx1.GetMean()
mean2 = hPx2.GetMean()
mean3 = hPx3.GetMean()

legend = ROOT.TLegend(0.6,0.6,0.9,0.9)
legend.AddEntry(hPx1, "601 MeV", "l")
legend.AddEntry(hPx2, "1003 MeV", "l")
legend.AddEntry(hPx3, "1405 MeV", "l")
legend.AddEntry(hPx1, "Mean: "+str(mean1), "l")
legend.AddEntry(hPx2, "Mean: "+str(mean2), "l")
legend.AddEntry(hPx3, "Mean: "+str(mean3), "l")
legend.Draw()

canvas.SaveAs("plots/hPx_3masses_pion.png")
display(Image(filename="plots/hPx_3masses_pion.png"))

```
