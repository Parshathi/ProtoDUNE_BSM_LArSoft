# Muon Muon versus Pion Pion for 1405MeV

## hPx

```python
import ROOT
from IPython.display import Image, display

f = ROOT.TFile("root_files//hepmc_1405MeV_pimu.root", "READ")
g = ROOT.TFile("root_files//hepmc_1405MeV_mumu.root", "READ")

folder1 = f.Get("generator")
folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)

pimu = folder1.Get("hPx")
mumu = folder2.Get("hPx")

mumu.SetLineColor(ROOT.kRed)
pimu.SetLineColor(ROOT.kBlue)

mumu.SetLineWidth(2)
pimu.SetLineWidth(2)

ROOT.gStyle.SetOptStat(0)

mumu.Draw()
pimu.Draw("same")


mumu.SetTitle("Muon-Muon and Pion-Muon x-Momentum Distribution")
pimu.SetTitle("Muon-Muon and Pion-Muon x-Momentum Distribution")

mumu.SetYTitle("Counts")
pimu.SetYTitle("Counts")

mean1 = mumu.GetMean()
mean2 = pimu.GetMean()

legend = ROOT.TLegend(0.7,0.7,0.9,0.9)
legend.AddEntry(mumu, r"$\mu- \mu$")
legend.AddEntry(pimu, r"$\pi - \mu$")
legend.AddEntry(mumu, "Mean: %.2f" % mean1, "l")
legend.AddEntry(pimu, "Mean: %.2f" % mean2, "l")
legend.Draw()

canvas.Draw()


canvas.SaveAs("plots/hpx_comp_1405.png")
display(Image(filename="plots/hpx_comp_1405.png"))


```

## Total Energy

```python
import ROOT

f = ROOT.TFile("root_files/hepmc_1405MeV_pimu.root", "READ")
g = ROOT.TFile("root_files/hepmc_1405MeV_mumu.root", "READ")

folder1 = f.Get("generator")
folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)


hE1 = folder1.Get("hEnergyTot")
hE1.SetLineColor(ROOT.kBlue)
hE1.SetLineWidth(2)
hE1.Draw()

hE2 = folder2.Get("hEnergyTot")
hE2.SetLineColor(ROOT.kRed)
hE2.SetLineWidth(2)
hE2.Draw("same")



hE1.SetTitle("Muon-Muon and Pion-Muon Energy Distribution")
hE2.SetTitle("Muon-Muon and Pion-Muon Energy Distribution")

hE1.SetYTitle("Counts")
hE2.SetYTitle("Counts")

mean1 = hE1.GetMean()
mean2 = hE2.GetMean()

legend = ROOT.TLegend(0.7,0.7,0.9,0.9)
legend.AddEntry(hE1, r"$\pi - \mu$")
legend.AddEntry(hE2, r"$\mu- \mu$")
legend.AddEntry(hE1, "Mean: %.2f" % mean1, "l")
legend.AddEntry(hE2, "Mean: %.2f" % mean2, "l")

legend.Draw()

canvas.Draw()
```
