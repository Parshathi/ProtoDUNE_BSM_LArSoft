# muon-muon vs pion-muon for 601 MeV

## hPx

```python
import ROOT
from IPython.display import Image, display

f = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_601MeV_pimu.root", "READ")
g = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_601MeV_mumu.root", "READ")

folder1 = f.Get("generator")
folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)

pimu = folder1.Get("hPx")
mumu = folder2.Get("hPx")

mumu.SetLineColor(ROOT.kRed)
pimu.SetLineColor(ROOT.kBlue)



ROOT.gStyle.SetOptStat(0)

mumu.Draw()
pimu.Draw("same")


mumu.SetTitle("Muon-Muon and Pion-Muon Momentum Distribution")
pimu.SetTitle("Muon-Muon and Pion-Muon Momentum Distribution")

mumu.SetYTitle("Counts")
pimu.SetYTitle("Counts")

legend = ROOT.TLegend(0.7, 0.7, 0.9, 0.9)
legend.AddEntry(mumu, "muon-muon")
legend.AddEntry(pimu, "pion-muon")
legend.Draw()

canvas.Draw()
canvas.SaveAs("histogram_comparison.png")
display(Image(filename="histogram_comparison.png"))

```

## Total Energy

```python
import ROOT

f = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_601MeV_pimu.root", "READ")
g = ROOT.TFile("/home/parshathi/Downloads/ProtoDUNE_BSM_LArSoft/hepmc_601MeV_mumu.root", "READ")

folder1 = f.Get("generator")
folder2 = g.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)


hE1 = folder1.Get("hEnergyTot")
hE1.SetLineColor(ROOT.kBlue)
hE1.SetLineWidth(2)
hE1.Draw()

hE2 = folder2.Get("hEnergyTot")
hE2.SetLineColor(ROOT.kRed)
hE2.SetLineWidth(2)
hE2.Draw("same")

legend = ROOT.TLegend(0.1, 0.7, 0.3, 0.9)
legend.AddEntry(hE1, "pion-muon")
legend.AddEntry(hE2, "muon-muon")

legend.Draw()

canvas.Draw()
```
