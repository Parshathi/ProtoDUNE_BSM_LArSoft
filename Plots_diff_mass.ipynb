# hPx for different masses - $\mu-\mu$

```python
# hPx for different masses, 601 MeV, 1003 MeV, 1045 MeV
import ROOT 
from IPython.display import Image, display
# open new TFile to read from
f = ROOT.TFile("root_files/hepmc_601MeV_mumu.root", "READ")
g = ROOT.TFile("root_files/hepmc_1003MeV_mumu.root", "READ")
h = ROOT.TFile("root_files/hepmc_1045MeV_mumu.root", "READ")
folder1 = f.Get("generator")
folder2 = g.Get("generator")
folder3 = h.Get("generator")

canvas =True
canvas = ROOT.TCanvas("canvas", "canvas", 900,550)

#plot the TObject hPx in the folder

hPx = folder1.Get("hPx")
hPx.SetLineColor(ROOT.kBlue)
hPx.SetLineWidth(2)
hPx.Draw()



hPx2 = folder2.Get("hPx")
hPx2.SetLineColor(ROOT.kRed)
hPx2.SetLineWidth(2)
hPx2.Draw("same")

ROOT.gStyle.SetOptStat(0)


hPx3 = folder3.Get("hPx")
hPx3.SetLineColor(ROOT.kGreen)
hPx3.SetLineWidth(2)
hPx3.Draw("same")



hPx.SetTitle("hPx for 3 masses "+r"$\mu - \mu$")
hPx2.SetTitle("hPx for 3 masses "+r"$\mu - \mu$")
hPx3.SetTitle("hPx for 3 masses "+r"$\mu - \mu$")

hPx.SetYTitle("Counts")
hPx2.SetYTitle("Counts")
hPx3.SetYTitle("Counts")

#legend for each mean, standard deviation
legend = ROOT.TLegend(0.7,0.7,0.9,0.9)
legend.AddEntry(hPx, "601 MeV", "l")
legend.AddEntry(hPx2, "1003 MeV", "l")
legend.AddEntry(hPx3, "1045 MeV", "l")
legend.Draw()




canvas.SaveAs("hPx_3masses.png")
display(Image(filename="hPx_3masses.png"))
```
